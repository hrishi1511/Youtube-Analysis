```python
# Task 5: Convert datatype of column published_at from object to pandas datetime
df['published_at'] = pd.to_datetime(df['published_at'], errors='coerce')
```

```python
# Task 6: Create a new column as 'published_month' using the column published_at
df['published_month'] = df['published_at'].dt.month
```

```python
df
```

```python
# Task 7: Replace the numbers in the column published_month as names of the months
month_names = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun', 7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'}
df['published_month'] = df['published_month'].map(month_names)
```

```python
# Task 8: Find the number of videos published each month and arrange the months in a decreasing order based on the video count
monthly_video_count = df['published_month'].value_counts().sort_values(ascending=False)
```
